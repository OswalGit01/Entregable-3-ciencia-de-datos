```python
import pandas as pd

df1= pd.read_csv("PDA.csv")
```

```python
df1.head(1)
```

```python
import seaborn as sns

#Primer grafico (diagrama de conteo)

sns.countplot(x="estadocum", data=df1, color= 'violet')
```

```python
# segunda grafica (Histograma)
import matplotlib.pyplot as plt

sns.histplot(df1["cantidad"])
plt.xlim(0, 1800)
plt.ylim(0, 800)
```

```python
#Tercer grafico (diagrama de tiras)

sns.stripplot(x="unidadmedida", y="consecutivocum", data=df1)
```

```python
# Cuarto grafico (diagrama de violin)

sns.violinplot(x="tiporol", y="concentracion", data=df1)
```

```python
# Quinto grafico (Mapa de calor)

corr = df1.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
```
