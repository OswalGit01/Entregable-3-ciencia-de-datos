```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#primera grafica (linea simple)
plt.plot([4, 8, 12, 16, 20, 24], [5, 10, 15, 20, 25, 30], color= 'green')
plt.title("Multiplos de 4 y 5")
plt.show()
```

```python
# Segunda grafica (Histograma)

datos = np.random.normal(0, 1, 1000)  # Distribución normal con media 0 y desviación estándar 1
plt.hist(datos, color = 'green', bins=20)
plt.title("Histograma de Datos Aleatorios")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

```python
import pandas as pd
datos1 = {
    'sexo': ['M', 'M', 'F', 'M', 'M', 'F', 'F', 'M', 'M', 'M'],
    'edad(años)': [20, 21, 24, 22, 17, 24, 15, 12, 19, 26],
    'altura(m)': [1.52, 1.63, 1.53, 1.82, 1.80, 1.78, 1.79, 1.72, 1.73, 1.66],
    'peso(kg)': [55, 67, 65, 68, 58, 66, 55, 61, 60, 59],}

df1 = pd.DataFrame(datos)
print(df1)
```

```python
# Grafica numero 3 (dispersoin)
x = [datos1['edad(años)'] ]
y = [datos1['peso(kg)'] ]
plt.scatter(x, y)
plt.title("Dispersión de peso por edad")
plt.xlabel("Edad cumplida")
plt.ylabel("Peso (kg)")
plt.show()
```

```python
# Cuarto grafico (Cajas y bigotes)
plt.boxplot(datos1['altura(m)'])
plt.title("Boxplot de altura")
plt.show()
```

```python
# Quinto gráfico (barras horizontales)

plt.barh(['Media Peso recien nacido', 'Media peso adultos', 'Media peso adultos mayores'], [7.2, 68.5, 56.5], color = 'red')
plt.title("Promedio de peso segun etapa de vida")
plt.show()
```
